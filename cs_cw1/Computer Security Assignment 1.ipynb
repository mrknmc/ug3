### part a

We use the Python library gmpy to verify the requirements for the El Gamal algorithm.

```python
import gmpy

p = gmpy.mpz(63689)
g = gmpy.mpz(14569)
k = gmpy.mpz(11636)
```

**Requirements:**

**1.** Number `p` needs to be prime. gmpy has a function `is_prime` which returns 2 when its argument is prime. Since `p` is prime, its greatest common divisor with g is clearly 1.

```python
2 == gmpy.is_prime(p) && 1 == gmpy.gcd(p, g)
```

**2.** Number `g` needs to be a generator. Thus it must be a member of the cyclic group $(Z_p)^*$ which in this case contains all the numbers from 2 smaller than `p` because `p` is prime. Moreover, it must be able to generate all of the numbers in this cyclic group when raised to the power of each of the elements in $(Z_p)^*$. Therefore, the set generated by `g` and the set of numbers from 2 to `p` - 1 should be equal: $\{g^1, g^2, ..., g^{p-1}\} = \{1, 2, ..., p - 1\}$.

```python
g in range(2, p) and set(pow(g, i, p) for i in range(1, p - 1)) == set(range(2, p))
```

**3.** Furthermore, the private key `k` needs to be from the set $\{1, ..., p - 2\}$ generated by the generator.

```python
k < p - 1
```

### part b

To find the decryption of the ciphertext we use `gmpy` again. Firstly, we are given c1 and c2. Secondly, to decrypt we use the decryption formula from the El Gamal algorithm $c_1^{-k} \cdot c_2 \pmod{p}$:

```python
c1, c2 = 7265, 44824
pow(c1, p - k - 1) * c2 % p
```
